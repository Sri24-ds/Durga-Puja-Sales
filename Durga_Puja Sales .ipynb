```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(15)
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Orders'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
df.rename(columns= {'Orders':' No_of_Order'})
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_counts = df['Occupation'].value_counts()

# create a pie chart
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.1f%%')

# add a title
plt.title('Occupations')

# show the plot
plt.show()
```

```python
sns.scatterplot(df['Age Group']),
df['Amount']
```

```python
state_counts = df['State'].value_counts()

# create a pie chart
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%')

# add a title
plt.title('STATE')

# show the plot
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()

# create a pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')

# add a title
plt.title('GENDER')

# show the plot
plt.show()
```

```python
sales_state = df.groupby(['Zone'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Zone',y= 'Orders')
```
